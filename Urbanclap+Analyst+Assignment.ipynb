```python
time = 8
def timeColumns(time):
    di = df_prov1['new_datetime_1']
    df=df_prov1['new_datetime']
    dci=pd.to_datetime('2017-09-01 0'+str(time)+':00:00')
    dcf=pd.to_datetime('2017-09-01 0'+str(time+1)+':00:00')
    df_prov1[str(time)] = np.where((df<dci) | (di>dcf),0 ,df_prov1['diff_seconds'])
    df_prov1[str(time)] = np.where((df_prov1['diff_seconds']>0)&((di>=dci)&(df<=dcf)),(df-di)/np.timedelta64(1,'s'),df_prov1[str(time)])
    df_prov1[str(time)] = np.where((df_prov1['diff_seconds']>0)&((di<=dci)&(df>=dcf)),3600,df_prov1[str(time)])
    df_prov1[str(time)] = np.where((df_prov1['diff_seconds']>0)&((dci<=df)&(df<=dcf)),(df-dci)/np.timedelta64(1,'s'),df_prov1[str(time)])
    df_prov1[str(time)] = np.where((df_prov1['diff_seconds']>0)&((dci<di)&(dcf>=di)),3600-(di-dci)/np.timedelta64(1,'s'),df_prov1[str(time)])
```

```python
def iterate(l, s, e):
    if s < 0 or e >= len(l) or s > e:
        return
    print(l[s])
    timeColumns(l[s])
    iterate(l, s + 1, e)
```
